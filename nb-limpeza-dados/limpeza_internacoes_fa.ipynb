# Limpeza: Casos de Febre Amarela

Importações

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import date
```

Leitura dos dados

```python
# 1984 a 2007 por processamento
url_int1 = 'https://github.com/itsGab/projeto-bootcamp-modulo-2-data-science/raw/main/dados-brutos/internacoes%20por%20uf%20segunda%20ano%20de%20processamento%20de%201984%20a%202007.csv'

# 2008 a 2021 por atendimento
url_int2 = 'https://github.com/itsGab/projeto-bootcamp-modulo-2-data-science/raw/main/dados-brutos/internacoes%20por%20uf%20segunda%20ano%20de%20atendimento%20de%202008%20a%202021.csv'
```

```python
df1 = pd.read_csv(url_int1, sep=';', encoding='latin-1', engine='python', 
                  decimal=',', thousands='.', skiprows=4, skipfooter=2)
df1
```

```python
df1 = df1.drop('Total', axis=1)
```

```python
df1 = df1.replace('-', 0).astype(int)
```

```python
df1.columns = df1.columns.str.lower()
```

```python
df1['ano processamento'] = df1['ano processamento'].map({ano: date(ano,1,1) for ano in df1['ano processamento']})
```

```python
df1
```

```python
df2 = pd.read_csv(url_int2, sep=';', encoding='latin-1', engine='python', 
                  decimal=',', thousands='.', skiprows=4, skipfooter=7)
```

```python
df2 = df2.drop('Total', axis=1)
df2 = df2.replace('-', 0).astype('int64')
df2.columns = df2.columns.str.lower()
df2['ano atendimento'] = df2['ano atendimento'].map({ano: date(ano,1,1) for ano in df2['ano atendimento']})
```

```python
df1_melt = df1.melt('ano processamento', var_name='uf', value_name='internacoes')
df2_melt = df2.melt('ano atendimento', var_name='uf', value_name='internacoes')
```

```python
df1.to_csv('internacoes_fa_por_ano_proce_pivot.csv')
df2.to_csv('internacoes_fa_por_ano_atend_pivot.csv')

df1_melt.to_csv('internacoes_fa_por_ano_proce_melt.csv')
df2_melt.to_csv('internacoes_fa_por_ano_atend_melt.csv')
```
